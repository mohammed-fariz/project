```python
import pandas as pd
from sklearn.experimental import enable_iterative_imputer  # Needed for IterativeImputer
from sklearn.impute import IterativeImputer
from sklearn.linear_model import LinearRegression

# Example DataFrame
# df = pd.read_csv("your_file.csv")

# Select numerical features (excluding target if necessary)
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Initialize Iterative Imputer with Linear Regression as estimator
imputer = IterativeImputer(estimator=LinearRegression(), 
                           max_iter=10, 
                           random_state=42)

# Fit and transform only the numerical columns
df[numerical_cols] = imputer.fit_transform(df[numerical_cols])

# Check if 'agent' has NaN filled
print(df['agent'].isna().sum())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset (replace with your path)

# ----------------------------
# SEPARATE NUMERICAL & CATEGORICAL COLUMNS
# ----------------------------

numerical_cols = df.select_dtypes(include=['int64', 'float64', 'int32']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

print("Numerical Columns:", numerical_cols)
print("Categorical Columns:", categorical_cols)

# ----------------------------
# EDA FOR NUMERICAL COLUMNS (Density Plot + Histogram)
# ----------------------------

for col in numerical_cols:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution / Density Plot of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

# ----------------------------
# EDA FOR CATEGORICAL COLUMNS (Bar Plot / Count Plot)
# ----------------------------

for col in categorical_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(x=df[col], order=df[col].value_counts().index)
    plt.title(f"Count Plot of {col}")
    plt.xticks(rotation=90)
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64', 'int32'])

# -----------------------------------------
# Heatmap: Correlation Matrix
# -----------------------------------------
plt.figure(figsize=(14, 10))
sns.heatmap(numeric_df.corr(), cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap of Hotel Booking Dataset")
plt.show()
```

```python
sns.pairplot(df, vars=numeric_df, hue="is_canceled", diag_kind="kde")
plt.show()
```
